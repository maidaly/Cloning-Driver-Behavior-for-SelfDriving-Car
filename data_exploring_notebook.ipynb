```python
import pandas as pd 
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import os
from utils import preprocess, random_flip, random_shadow, random_brightness, random_translate
plt.style.use('ggplot')
%matplotlib inline
```

```python
data_df = pd.read_csv(os.path.join('data', 'driving_log.csv'), names=['center', 'left', 'right', 'steering', 'throttle', 'reverse', 'speed'])
```

```python
print ("driving_log.csv")
data_df.head()
```

```python
steering_angle = data_df["steering"].values
data_df["steering"].describe()
```

**Steering angle distrbution**

```python
fig = plt.figure(figsize=(12,6))
plt.hist(steering_angle,alpha=0.60,bins=25,ec='w')
plt.ylabel("Frequancy")
plt.title('Histogram of Steering angle')
```

The high frequancy of 0 can lead to bias the model to 0. we can avoid this by making copies of non zeros and try to apply some Augmentation function to have better distrbution.

**Preprocessing**

In Preprocessing 3 functions are applied to the image ( crop, resize, rgb2yuv). The crop function removes unnecessary parts from the image like the sky and the front part of the car as they hold no information. The resize function resizes the image to the input shape used by the network model (66, 200). The rgb2yuv converts the image from RGB to YUV (This is what the NVIDIA model does)

```python
images= data_df[['left','center', 'right']].values
```

```python
list_of_titles = ['left','center', 'right']
list_of_images = images[100]
fig = plt.figure(figsize=(20,7))
column = 1
print("Sample af training images and the crosponding preprocessing")
for i in range(len(list_of_images)):
        plt.subplot(2, 3, column)
        image = np.array(plt.imread(list_of_images[i]))
        plt.imshow(image)
        plt.imsave("original_image_"+list_of_titles[i]+".png",image)
        plt.axis('off')
        plt.title("Original image" +"\n"+list_of_titles[i])
        plt.subplot(2, 3, column+3)
        plt.imshow(preprocess(image))
        plt.imsave("preprocess_image_"+list_of_titles[i]+".png",preprocess(image))
        plt.axis('off')
        plt.title("Preprocessing image" +"\n"+list_of_titles[i])
        column += 1
```

 ### Image Augmentation
 
The image augmentation is done by applaying 4 functions randomly to the images the 4 functions are:

***random_flip*** : Flipping the copies of image horizontally, and negating the steering angles. this function tries to add some                     balance to the steering_angle distribution.

***random_translate*** : Randomly translate image with steering angle adjustment (0.002 per pixel shift).

***random_shadow*** : Randomly added shadows to fit diffrent conditions.

***random_brightness*** : Randomly change brightness to fit diffrent conditions (lighter or darker).

```python
list_of_titles = ['left','center', 'right']
list_of_images = images[100]
steering = steering_angle[100]
fig = plt.figure(figsize=(20,25))
column = 1
print("Sample af augmentation functions and the crosponding preprocessing")
for i in range(len(list_of_images)):
        plt.subplot(6, 3, column)
        image = np.array(plt.imread(list_of_images[i]))
        plt.imshow(image)
        plt.axis('off')
        plt.title("Original image , Steering angle: {}".format(np.round(steering,3)) +"\n"+list_of_titles[i])
        column += 1
        
# flip image        
for i in range(len(list_of_images)):
        plt.subplot(6, 3, column)
        image = np.array(plt.imread(list_of_images[i]))
        processed_image, new_steering = random_flip(image,steering, force_flip =True)
        plt.imshow(processed_image)
        plt.axis('off')
        plt.title("fliping image , New steering angle: {}".format(np.round(new_steering,3)) +"\n"+list_of_titles[i])
        plt.imsave("flip_"+list_of_titles[i]+".png",processed_image)        
        column += 1

# translate image        
for i in range(len(list_of_images)):
        plt.subplot(6, 3, column)
        image = np.array(plt.imread(list_of_images[i]))
        processed_image, new_steering = random_translate(image,steering, range_x= 50, range_y= 50)
        plt.imshow(processed_image)
        plt.axis('off')
        plt.title("translating image , New steering angle: {}".format(np.round(new_steering,3)) +"\n"+list_of_titles[i])
        plt.imsave("translate_"+list_of_titles[i]+".png",processed_image)
        column += 1

# add random_shadow         
for i in range(len(list_of_images)):
        plt.subplot(6, 3, column)
        image = np.array(plt.imread(list_of_images[i]))
        processed_image = random_shadow(image)
        plt.imshow(processed_image)
        plt.axis('off')
        plt.title("shadowing image , Steering angle: {}".format(np.round(steering,3)) +"\n"+list_of_titles[i])
        plt.imsave("shadow_"+list_of_titles[i]+".png",processed_image)
        column += 1
        
# add random_brightness
for i in range(len(list_of_images)):
        plt.subplot(6, 3, column)
        image = np.array(plt.imread(list_of_images[i]))
        processed_image = random_brightness(image)
        plt.imshow(processed_image)
        plt.axis('off')
        plt.title("random_brightnes image , Steering angle: {}".format(np.round(steering,3)) +"\n"+list_of_titles[i])
        plt.imsave("brightness_"+list_of_titles[i]+".png",processed_image)
        column += 1
```
